# Vectorizing and Topic Modeling
In this notebook I will be doing various vectorizing and topic modeling of our cleaned text data. Some approaches I will be using are Latent Dirilhect Alloction, VADER sentiment anlaysis for word sentiment and review sentiment and a custom Word2Vec model that has only 2 embedding dimensions so we can actually graph the word vectors in 2-D space. I will be using a Tf-idf vectorizer for the vectorizing of our data for most of the applications.

```python
from nltk.corpus import stopwords 
from nltk.stem.wordnet import WordNetLemmatizer
import pandas as pd
import string
from sklearn.decomposition import LatentDirichletAllocation, TruncatedSVD
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import gensim
from gensim import corpora
import numpy as np
import spacy
import textacy
import pickle
import time
import matplotlib.pyplot as plt
from nltk.sentiment.vader import SentimentIntensityAnalyzer

%matplotlib inline
```

```python
with open('./tokenized_rest_reviews.pkl', 'rb') as f:
    rests = pickle.load(f)
```

Because we tokenized the text already and wont need to preprocess or tokenize it again, which TF-IDF can do automatically, we are passing a basic function that will essentially not do anything but allow the vectorizer to run over the already tokenized text.

```python
def identity(doc):
    return doc

tf_vec = TfidfVectorizer(analyzer='word', tokenizer=identity, preprocessor=identity, 
                             token_pattern=None, strip_accents='ascii', min_df=5, max_df=0.95,
                             max_features=10000, use_idf=True, sublinear_tf=True)
```

```python
doc_term_matrix = tf_vec.fit_transform((doc for doc in rests))
```

```python
features = tf_vec.get_feature_names()
```

```python
def display_topics(model, feature_names, num_words):
    for ix, topic in enumerate(model.components_):
        print(f'Topic {ix}:')
        words = [feature_names[i] for i in topic.argsort()[:-num_words - 1:-1]]
        print('\n'.join(words))
        print()
```

Fitting a LDA model with 20 topics

```python
lda = LatentDirichletAllocation(n_components = 20, random_state=42)
lda.fit(doc_term_matrix)
```

```python
display_topics(lda, features, 5)
```

Defining 2 functions to allow visualization of key components of the LSA model.

```python
# Credit to George Dewey from DSI Cohort before mine for these funtions and the termite plot code below)
def top_topic_docs(corpus, model, doc_topic_matrix, n_topics):
    for topic_idx, top_docs in model.top_topic_docs(doc_topic_matrix, topics=n_topics, top_n=1):
        print(topic_idx)
        for j in top_docs:
            print(corpus[j])
```

```python
def print_topics(model, n_topics, n_terms):
    for topic_idx, top_terms in model.top_topic_terms(tf_vec.get_feature_names(), topics=n_topics, top_n=n_terms):
        print('topic', topic_idx, ':', '   '.join(top_terms))
```

```python
%%time
lsa_model = TruncatedSVD(n_components=100, random_state=1)
lsa_model.fit(doc_term_matrix)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(100), np.cumsum(lsa_model.explained_variance_ratio_), color='blue')
plt.title('% variance explained by LSA topics', size=16)
plt.xlabel('Number of topics', size=12)
plt.ylabel('Variance explained', size=12)
plt.savefig('./visuals/LSA_var.png')
```

Instantiating a textacy LSA model for topic modeling and termite plot graphing

```python
%%time
lsa_model = textacy.TopicModel('lsa', n_topics=100, random_state=1)
lsa_model.fit(doc_term_matrix)

lsa_doc_topic_matrix = lsa_model.transform(doc_term_matrix)
```

```python
print_topics(lsa_model, n_topics=[num for num in range(1, 20)], n_terms=5)
```

```python
lsa_model.termite_plot(doc_term_matrix,
                       tf_vec.get_feature_names(), 
                       topics=[num for num in range(1, 20)],
                       n_terms=15, 
                       highlight_topics=[1, 3, 10, 11, 13, 19],
                       save='./visuals/rest_termite_plot.png')
```

### Vader Sentiment Analysis on Tf-idf vectorizer
Below I wanted to run a VADER sentiment analysis on our tf-idf vectorizer. I also will be running VADER on a Word2Vec model in the next notebook to see any major differences in how the sentiment analysis classifies words based on the vecotrizer.

```python
train_set = pd.DataFrame(doc_term_matrix.todense(), columns =features)
```

```python
sia = SentimentIntensityAnalyzer()

dicts = []

for word in train_set.columns:
    scores = sia.polarity_scores(word)
    scores['word'] = word
    dicts.append(scores)

df_vader = pd.DataFrame(dicts)

df_vader.info()
```
